### 리스트 컴프리헨션
- 기존의 리스트 사용법을 좀 더 간결하고 효율적으로 사용위해서

````python
# 기존방식 1
result1 = []
for i in range(10):
    result1.append(i)
````

````python
result1
````

````python
# 기존방식 2
result1.clear()

for i in range(2,40,2):
    result1.append(i)
````

````python
result1
````

````python
# 리스트 컴프리헨선 방식 1
result2 = [i for i in range(10)]
````

````python
result2
````

````python
# 리스트 컴프리헨션 방식 2
result2 = [i for i in range(2,40,2)]
````

````python
result2
````

````python
result2 = [i*2 for i in range(1,20)]
result2
````

#### Python 리스트와 NumPy 배열의 차이

- Python 리스트는 한 요소끼리 더하는 작업을 반복문 사용해서 처리

````python
# Python 리스트
a = [1,2,3]
b = [4,5,6]

res_list = [a[i] + b[i] for i in range(3)]  # 리스트 컴프리헨션
res_list
````

````python
res_list = []

for i in range(3): # 일반 반복문
    res_list.append(a[i]+b[i])

res_list
````
